# Ny tittel
Tittel kan lages ved å bruke Markdown og #

```python
import numpy as np
import matplotlib.pyplot as plt 

theta = np.linspace(0, 2*np.pi, 500)[:,None]
r = np.linspace(1, 5, 5)[None,:]

rr = (r*np.cos(theta), r*np.sin(theta))
plt.plot(*rr)
plt.axis('equal')
# Trenger ikke plt.show()
```

```python
fig = plt.figure()
gs = fig.add_gridspec(2,2)
ax1 = fig.add_subplot(gs[0,0], projection='polar')
ax1.plot(*rr)
ax2 = fig.add_subplot(gs[0,1], projection='3d')
ax3 = fig.add_subplot(gs[1,:])
ax3.axis('equal')
ax3.plot(*rr)
```
